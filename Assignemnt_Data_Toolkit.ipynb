**Q1.What is NumPy, and why is it widely used in Python?**
- **Ans** NumPy (Numerical Python) is a powerful open-source Python library used for numerical and scientific computing. It provides support for large, multi-dimensional arrays and matrices, along with a wide range of high-performance mathematical functions to operate on them efficiently. Its ability to handle complex mathematical operations with simple syntax makes it a core tool in the Python ecosystem.


**Q2.How does broadcasting work in NumPy?**
- **Ans.**In NumPy, broadcasting is a mechanism that allows arrays of different shapes to be used together in arithmetic operations without explicitly copying or reshaping them. When two arrays are operated on, NumPy compares their shapes element-wise, starting from the trailing (rightmost) dimensions. If the dimensions are equal, or if one of them is 1, they are considered compatible; otherwise, broadcasting fails. The smaller array is virtually "stretched" (without making copies) to match the larger array’s shape, enabling element-wise operations. This makes computations more efficient, memory-friendly, and concise, avoiding the need for writing loops manually.

**Q3.What is a Pandas DataFrame?**
- **Ans**A **Pandas DataFrame** is a two-dimensional, tabular data structure in Python, similar to a spreadsheet or SQL table. They come with powerful built-in methods for handling missing values, grouping, merging, reshaping, and performing statistical or mathematical operations. Because of their flexibility and ease of use, DataFrames are one of the most commonly used tools for data analysis and manipulation in Python.


**Q4.Explain the use of the groupby() method in Pandas?**
- **Ans.**The **`groupby()`** method in Pandas is used to split data into groups based on the values of one or more columns, and then perform operations such as aggregation, transformation, or filtering on each group. It follows a **split-apply-combine** strategy: first, the data is divided into groups; second, a function (like `sum()`, `mean()`, `count()`, etc.) is applied to each group; and finally, the results are combined into a new DataFrame or Series. This makes `groupby()` especially useful for analyzing and summarizing large datasets, such as finding the average sales per region, total marks per student, or count of items in each category.

**Q6.What are the differences between NumPy arrays and Python lists?**
- *Ans.* The main difference between **NumPy arrays** and **Python lists** is that NumPy arrays are optimized for numerical and scientific computing, while Python lists are general-purpose containers. NumPy arrays are stored in a continuous block of memory, making them more efficient, faster, and capable of handling large datasets with vectorized operations, whereas Python lists store elements as objects, leading to more memory usage and slower performance for numerical tasks. Additionally, NumPy arrays support advanced mathematical operations, broadcasting, and multi-dimensional data, which regular Python lists do not provide directly.

**Q8.What does the term “vectorized operation” mean in NumPy?**
- **Ans** In NumPy, a **vectorized operation** means performing arithmetic or mathematical operations on entire arrays (vectors, matrices) at once, without writing explicit loops. Instead of iterating through elements one by one, NumPy applies the operation simultaneously across all elements using optimized C-based implementations. This makes computations much faster, more concise, and easier to read compared to traditional Python loops.

**Q17.Explain the difference between apply() and map() in Pandas?**
- **Ans** In Pandas, **`map()`** is mainly used with Series to apply a function element-wise, often for simple transformations or value mappings. On the other hand, **`apply()`** is more flexible and can be used with both Series and DataFrames. For Series, it works similarly to `map()`, but for DataFrames, it can apply a function along rows or columns. In short, `map()` is best for element-wise operations on a single column, while `apply()` is more powerful for applying custom functions across rows, columns, or entire DataFrames.

**Q21.Why is NumPy’s array slicing faster than Python’s list slicing?**
- **Ans.** NumPy’s array slicing is faster than Python’s list slicing because **NumPy arrays are stored in a contiguous block of memory** with a fixed data type, while Python lists are collections of pointers to objects scattered in memory. When slicing a NumPy array, it creates a **view** (not a copy) that references the same underlying data without duplicating it, making the operation very efficient. In contrast, slicing a Python list always creates a new list and copies the elements, which takes more time and memory. This memory-efficient design is a key reason NumPy is preferred for numerical computing.

```python
#Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Math': [90, 85, 78, 92, 88],
    'Science': [85, 80, 70, 95, 89],
    'English': [75, 80, 85, 70, 78],
    'History': [80, 70, 88, 85, 82]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Q11. Perform matrix multiplication using NumPy?
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Matrix multiplication
result = np.dot(A, B)   # or A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result of Matrix Multiplication:\n", result)
```

```python
#Q13.Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'X': [1, 2, 3, 4, 5, 6],
    'Y': [10, 20, 25, 30, 35, 40],
    'Z': [5, 15, 10, 25, 20, 30],
    'Category': ['A', 'A', 'B', 'B', 'C', 'C']
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z',
                    color='Category',  # Different colors by category
                    size='Z',          # Point size based on Z values
                    title="3D Scatter Plot Example")

fig.show()
```
